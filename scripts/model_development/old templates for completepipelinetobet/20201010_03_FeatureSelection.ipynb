```python
# Function for Correlation Heatmap
def heatmap(df):
    
    # Correlation Coefficient Calculation
    df_corr = df.corr(method='spearman')
    
    # Adjustment of Matrix Dimensions
    matrix = np.triu(df_corr)
    
    # Defining Font Size
    sns.set(font_scale=0.65)
    
    # Heatmap
    return sns.heatmap(round(df_corr, 1), annot=True, fmt='.1g', square=True, mask=matrix, vmin=-1, vmax=1, center=0, cmap='RdYlGn', annot_kws={"fontsize":5})
```

```python
# Function to Compute an Empirical Cumulative Distribution Function (ECDF)
def ecdf(data):

    # Number of Data Points
    n = len(data)

    # Data on X-Axis for the ECDF
    x = np.sort(data)

    # Data on Y-Axis for the ECDF
    y = np.arange(1, n + 1) / n

    return x, y
```

```python
# Seed random number generator
np.random.seed(20)

# Take 10,000 samples out of the binomial distribution: n_defaults
n_defaults = np.random.binomial(n=50, p=0.5, size=100000)

# Compute bin edges: bins
bins = np.arange(min(n_defaults), max(n_defaults) + 1.5) - 0.5

# Histogram Plot
# Background color
_ = plt.figure(figsize=(10, 4))
_ = plt.axvspan(10, 16, facecolor='#B22222', alpha=0.5)
_ = plt.axvspan(16, 34, facecolor='#FFEC8B', alpha=0.5)
_ = plt.axvspan(34, 40, facecolor='#228b22', alpha=0.5)
_ = plt.hist(n_defaults, normed=True, bins=bins)

# Margins
_ = plt.margins(0.02)

# Axis Labels
_ = plt.xlabel('Number of Importance Hits')
_ = plt.ylabel('Binomial distribution | n=50, p=0.5')

# Save Plot as JPEG
plt.savefig('Graphics/Boruta.jpg', dpi=300)

plt.show()
```
